```python
df1.replace({'NAME_CONTRACT_TYPE':{'Cash loans': 0, 'Revolving loans': 1}}, inplace=True)
df_test1.replace({'NAME_CONTRACT_TYPE':{'Cash loans': 0, 'Revolving loans': 1}}, inplace=True)

df1.replace({'CODE_GENDER':{'M': 0, 'F': 1}}, inplace=True)
df_test1.replace({'CODE_GENDER':{'M': '0', 'F': '1'}}, inplace=True)

df1.replace({'FLAG_OWN_CAR':{'N': 0, 'Y': 1}}, inplace=True)
df_test1.replace({'FLAG_OWN_CAR':{'N': 0, 'Y': 1}}, inplace=True)

df1.replace({'FLAG_OWN_REALTY':{'N': 0, 'Y': 1}}, inplace=True)
df_test1.replace({'FLAG_OWN_REALTY':{'N': 0, 'Y': 1}}, inplace=True)
```

```python
df1.drop('NAME_INCOME_TYPE', axis=1, inplace=True)
df_test1.drop('NAME_INCOME_TYPE', axis=1, inplace=True)

df1.drop('NAME_EDUCATION_TYPE', axis=1, inplace=True)
df_test1.drop('NAME_EDUCATION_TYPE', axis=1, inplace=True)

df1.drop('NAME_FAMILY_STATUS', axis=1, inplace=True)
df_test1.drop('NAME_FAMILY_STATUS', axis=1, inplace=True)

df1.drop('NAME_HOUSING_TYPE', axis=1, inplace=True)
df_test1.drop('NAME_HOUSING_TYPE', axis=1, inplace=True)

df1.drop('OCCUPATION_TYPE', axis=1, inplace=True)
df_test1.drop('OCCUPATION_TYPE', axis=1, inplace=True)

df1.drop('WEEKDAY_APPR_PROCESS_START', axis=1, inplace=True)
df_test1.drop('WEEKDAY_APPR_PROCESS_START', axis=1, inplace=True)

df1.drop('ORGANIZATION_TYPE', axis=1, inplace=True)
df_test1.drop('ORGANIZATION_TYPE', axis=1, inplace=True)

df1.drop('HOUSETYPE_MODE', axis=1, inplace=True)
df_test1.drop('HOUSETYPE_MODE', axis=1, inplace=True)

df1.drop('WALLSMATERIAL_MODE', axis=1, inplace=True)
df_test1.drop('WALLSMATERIAL_MODE', axis=1, inplace=True)

df1.drop('EMERGENCYSTATE_MODE', axis=1, inplace=True)
df_test1.drop('EMERGENCYSTATE_MODE', axis=1, inplace=True)

df1.drop('FONDKAPREMONT_MODE', axis=1, inplace=True)
df_test1.drop('FONDKAPREMONT_MODE', axis=1, inplace=True)
```

```python
df = df1.dropna()
df_test = df_test1.dropna()
```

```python
df.head(100)
```

```python
X = df.iloc[:, 9:].values
y = df.iloc[:, 8].values

X_test = df_test.iloc[:, 8:].values
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.ensemble import RandomForestRegressor
#ランダムフォレスト
rfc = RandomForestRegressor(max_depth=15, min_samples_leaf=2, n_estimators=100, n_jobs=-1, random_state=42)
rfc.fit(X_train, y_train)
```

```python
print('Train Score: {}'.format(round(rfc.score(X_train, y_train), 3)))
print(' Test Score: {}'.format(round(rfc.score(X_valid, y_valid), 3)))
```

```python
param_grid = {'max_depth': [3, 5, 7],
              'min_samples_leaf': [1, 2, 4]}

for max_depth in param_grid['max_depth']:
    for min_samples_leaf in param_grid['min_samples_leaf']:
        rfc_grid = RandomForestRegressor(max_depth=max_depth, min_samples_leaf=min_samples_leaf, 
                                            n_estimators=100, n_jobs=-1, random_state=42)
        rfc_grid.fit(X_train, y_train)
        print('max_depth: {}, min_samples_leaf: {}'.format(max_depth, min_samples_leaf))
        print('    Train Score: {}, Test Score: {}'.format(round(rfc_grid.score(X_train, y_train), 3),
                                                           round(rfc_grid.score(X_valid, y_valid), 3)))
```

```python
rfc_gs = GridSearchCV(RandomForestRegressor(n_estimators=100, n_jobs=-1, random_state=42), param_grid, cv=5)
rfc_gs.fit(X, y)

print('Best Parameters: {}'.format(rfc_gs.best_params_))
print('CV Score: {}'.format(round(rfc_gs.best_score_, 3)))
```

```python
pred = rfc_gs.best_estimator_.predict(X_test)
```
